```python
df = df.drop(columns=["unnamed_col"])
df
```

```python
df = df.rename(columns={"Product_name": "Product Name", "cost": "Cost", "model": "Model", "entry_date": "Entry Date"})
df
```

```python
df['Brand'] = df['Brand'].str.title()
df['Origin'] = df['Origin'].str.title()
df['Product Name'] = df['Product Name'].str.title()
df
```

```python
df.info()
```

```python
df['Cost_inr'] = (
    df['Cost']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
)
df['Cost_inr'] = pd.to_numeric(df['Cost_inr'], errors="coerce")
df.loc[df['Cost'].astype(str).str.contains('$', regex=False), 'Cost_inr'] *= 89
df
# df.info()
```

```python
df = df.drop(columns=("Cost"))
df
```

```python
df['Cost_inr'] = df['Cost_inr'].fillna(df['Cost_inr'].median())
df
```

```python
df['Entry Date'] = df['Entry Date'].astype('datetime64[ms]')
df
```

```python
df.info()
```

```python
df['Model'] = df['Model'].str.replace(",", "-")
df['Model'] = df['Model'].str.replace(":", "-")
df
```

```python
df['Model'] = df['Model'].str.title()
df['Id'] = df['Id'].str.upper()
df
```

```python
df[["Model No", "Model"]] = df['Model'].str.split("-", expand=True)
df
```
